Esta sección del Laboratorio titulada `Estadística Descriptiva` tiene por objetivo la entrega de: 

+ Un **DataFrame** con toda la información histórica de operaciones y la actividad de trading. Este contenedor incluye datos de las operaciones realizadas por cada uno de los traders; como la fecha de apertura y cierre de sus posiciones, los instrumentos sobre los cuales se operó, el tamaño de la posición (lotaje) en cada trade, los tipo de operación (compra o venta) y el beneficio obtenido en cada una.

La primera llave retorna un *DataFrame* con estadísticas básicas, como el número de operaciones ganadoras y perdedoras, y algunos ratios de efectividad y proporción, que miden el desempeño del trader. La segunda llave retorna un *DataFrame* que muestra el ratio de efectividad de las operaciones realizadas con cada instrumento operado.
